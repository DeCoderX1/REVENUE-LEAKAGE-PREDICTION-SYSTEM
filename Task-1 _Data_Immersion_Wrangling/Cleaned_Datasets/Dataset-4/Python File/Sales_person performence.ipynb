```python
query="""
SELECT
    sp.salesperson_id,
    sp.salesperson_name,
    sp.region,

    COUNT(so.order_id) AS total_orders,

    ROUND(
        SUM(
            CASE
                WHEN (so.list_price * (1 - dp.max_discount_pct / 100)) - i.invoice_price > 0
                THEN (so.list_price * (1 - dp.max_discount_pct / 100)) - i.invoice_price
                ELSE 0
            END
        ),
        2
    ) AS total_leakage,

    ROUND(AVG(so.approved_discount_pct), 2) AS avg_discount_given

FROM sales_orders so
JOIN salespersons sp
    ON so.salesperson_id = sp.salesperson_id
JOIN invoices i
    ON so.order_id = i.order_id
JOIN discount_policy dp
    ON so.product = dp.product

GROUP BY
    sp.salesperson_id,
    sp.salesperson_name,
    sp.region

ORDER BY total_leakage DESC;
"""
```

```python
salesperson_perf = pd.read_sql(query, conn)
salesperson_perf.to_csv("salesperson_leakage_performance.csv", index=False)
salesperson_perf.head()
```
